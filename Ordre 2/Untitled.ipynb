# 1D_statioannaire_ordre2_negatif

```python
import numpy as np
import matplotlib.pyplot as plt 
ld=-100
p=1.9389
d=90
n=100
dx=1/n
l=np.zeros(n,dtype=np.double)
for i in range(1,n):
    l[i]=l[i-1]+(1/n)
m=np.eye(n,n)
for i in range(2,n):
    m[i,i-1]=-(2+(ld*dx*dx))
    m[i,i-2]=1
```

# Analyse du problème

```python
c=np.zeros((n,1))
c[0]=p
c[1]=p+(d*dx)
if (np.linalg.det(m)!=0):
    q=np.linalg.inv(m)
    s=np.dot(q,c)
else:
    print("Le problème est sans solution")
    print("le determinant de m est nul")
```

# Plot du problème

```python
x=np.linspace(0,1,1000000)
s_negatif=np.sqrt(-ld)
alpha_negatif=p
beta_negatif=d/s_negatif
y_negatif=(alpha_negatif*np.cos(s_negatif*x))+(beta_negatif*np.sin(s_negatif*x))

plt.scatter(l, s, label='Fonction une dimension',marker='*')  # 'label' is optional and provides a label for the curve in the legend
plt.plot(x,y_negatif,label="Solution generale",color='red')
plt.xlabel('X-axis Label')  # Label for the x-axis
plt.ylabel('Y-axis Label')  # Label for the y-axis
plt.title('2D Curve Plot')  # Title of the plot
plt.legend()  # Show legend based on the 'label' parameter in the plot function
plt.grid(True)  # Show grid in the plot
plt.show() 
```
