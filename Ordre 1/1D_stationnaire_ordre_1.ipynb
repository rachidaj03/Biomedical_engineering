# Poser le problème


```python
import numpy as np
import matplotlib.pyplot as plt 
ld=-3
a=9
n=100
dx=1/n
l=np.zeros(n,dtype=np.double)
for i in range(1,n):
    l[i]=l[i-1]+(1/n)
m=np.eye(n,n)
for i in range(0,n-1):
    m[i+1,i]=-(1+(ld*dx))
```

# Analyse du problème

```python
c=np.zeros((n,1))
c[0]=a
if (np.linalg.det(m)!=0):
    q=np.linalg.inv(m)
    s=np.dot(q,c)
else:
    print("Le problème est sans solution")
    print("le determinant de m est nul")
```

# Plot du problème

```python
x=np.linspace(0,1,10000)
y=a*np.exp(ld*x)
plt.plot(l, s, label='Fonction une dimension')  # 'label' is optional and provides a label for the curve in the legend
plt.plot(x,y,label="Exponentielle")
plt.xlabel('X-axis Label')  # Label for the x-axis
plt.ylabel('Y-axis Label')  # Label for the y-axis
plt.title('2D Curve Plot')  # Title of the plot
plt.legend()  # Show legend based on the 'label' parameter in the plot function
plt.grid(True)  # Show grid in the plot
plt.show() 
```
