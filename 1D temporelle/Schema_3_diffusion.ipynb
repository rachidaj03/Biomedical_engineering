```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np

#Paramètres du problème en concentration
c_0=1000
sigma=1000
c=0.01

#Paramètres de discrétisation de ©l'espace x
n=100
dx=1/n
x=np.zeros(n,dtype=np.double)
for i in range(1,n):
    x[i]=x[i-1]+(1/n)
#Paramètres de discrétisation du temps t
m=100
dt=1/m
t=np.zeros(m,dtype=np.double)
for i in range(1,m):
    t[i]=t[i-1]+(1/m)
#Paramètre alpha 
s=(c*dt)/(2*dx*dx)
#Matrice des conditions initiales
B_0=np.zeros(n,dtype=np.double)
for i in range(0,n):
    B_0[i]=x[i]

#Matrice des composantes temporelles
sol=np.zeros((m, n))
for i in range(0,n):
    sol[0,i]=B_0[i]
    
#Matrice C et D
C=np.eye(n)
D=np.eye(n)
for i in range(0,n):
    for j in range(0,n):
        if(j==i+1):
            C[i,j]=-s
        elif(j==i):
            C[i,j]=1+(2*s)
        elif(j==i-1):
            C[i,j]=-s
for i in range(0,n):
    for j in range(0,n):
        if(j==i+1):
            D[i,j]=s
        elif(j==i):
            D[i,j]=1-(2*s)
        elif(j==i-1):
            D[i,j]=s
C=np.linalg.inv(C)
A=np.dot(C,D)
for i in range(0,m):
    power = np.linalg.matrix_power(A, i)
    result=np.dot(power,B_0)
    for j in range(0,n):
        sol[i,j]=result[j]

#Matrice de la solution en fonction de t 
def f(x, t):
    return np.cos(x+(c*t))
```

```python
def create_3d_plot():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    X1, Y1 = np.meshgrid(x,t)
    result = f(X1, Y1)
    ax.plot_surface(X1, Y1, sol, cmap='viridis')
    #ax.plot_surface(X1, Y1, result, cmap='inferno')
    return fig
```

```python
root = tk.Tk()
root.title("Tracé 3D interactif")

# Créez le tracé 3D
fig = create_3d_plot()

# Créez un canevas Tkinter pour afficher le tracé
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack()

# Fonction pour effectuer une rotation du tracé en réponse à la souris
def on_mouse_move(event):
    if event.button == 'up':
        fig.canvas.release(event)
    else:
        fig.canvas.motion_notify(event)
def on_mouse_scroll(event):
    if event.button == 'up':
        scale = 1.1
    else:
        scale = 1 / 1.1

    ax = fig.gca(projection='3d')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    zlim = ax.get_zlim()
    ax.set_xlim(
        (event.xdata - xlim[0]) * scale + xlim[0],
        (event.xdata - xlim[1]) * scale + xlim[1]
    )
    ax.set_ylim(
        (event.ydata - ylim[0]) * scale + ylim[0],
        (event.ydata - ylim[1]) * scale + ylim[1]
    )
    ax.set_zlim(
        (event.ydata - zlim[0]) * scale + zlim[0],
        (event.ydata - zlim[1]) * scale + zlim[1]
    )

canvas.mpl_connect('motion_notify_event', on_mouse_move)
canvas.mpl_connect('button_release_event', on_mouse_move)
canvas.mpl_connect('scroll_event', on_mouse_scroll)
# Lancez l'application Tkinter
root.mainloop()
```
