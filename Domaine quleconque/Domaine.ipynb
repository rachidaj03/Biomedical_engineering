```python
import matplotlib.pyplot as plt
import numpy as np

# Créer des coordonnées x et y pour les points du maillage
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
x, y = np.meshgrid(x, y)

# Aplatir les tableaux pour obtenir une liste de coordonnées
x_flat = x.flatten()
y_flat = y.flatten()

# Créer une figure
plt.figure()

# Utiliser la fonction scatter pour représenter les points du maillage
plt.scatter(x_flat, y_flat, marker='o', color='blue')

# Ajouter des étiquettes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Afficher le graphique
plt.show()
```
